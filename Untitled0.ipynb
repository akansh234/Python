```python
num = int(input("Enter the number"))

num_str = str(num)
power = len(num_str)

total = sum(int(digit) ** power for digit in num_str)

if num == total:
    print(f"{num} is an Armstrong number")
else:
    print(f"{num} is NOT an Armstrong number")
```
